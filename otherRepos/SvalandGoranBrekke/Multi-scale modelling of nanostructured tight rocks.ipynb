# A flow of water in a portlandite system

I wish to make a current through surfaces of portlandite to measure the net flux through the nanostructure pore. From this we should be able through Darcy's law, to estimate the permeability of the thight rock.


\begin{equation}
Q = \frac{-\kappa A}{\mu }\frac{\Delta P}{\Delta x}
\end{equation}

Where Q is the volume flux of the material passing by the area $ A $, $ \mu $ is its viscosity (Pa$\cdot $s). And the flux is induced by a pressure gradient $ \Delta P $ over a distance $ \Delta x $.

## Laminar flow

Laminar flow is when we asume the fluid to flow in parallell layers with no disruption between the layers, e.g. no eddies (turbulence) in the flow. This is a good approximation at low velocities compared to the viscosity of the fluid. A fluid with high viscosity has a higher shear stress tensor, and flows more slowly than a fluid with low viscosity. A shear stress between fluid particles is present in a viscous fluid, and is defined as the shear force applied to the shear surface of the fluid particle. But since we dont want to operate with a fluid particles surface (hard to know its size), and we are more interested in the velocity profile across the direction of flow, a coefficient of viscosity is introduced, and rewrite the shear stress into:
\begin{equation}
\tau = \frac{F}{A} = \mu \frac{dv_x}{dl}
\end{equation}
Where $ dl $ is the distance between the fluid particles, and $ dv_x $ is the relative velocity of the fluid particles. We notice that the coefficient of viscosity $ \mu $ has units of $ [Pa \cdot s ] $. A flow in x-direction is driven by a pressure difference $ \Delta P $, the net forces acting on a fluid particle (asuming laminar flow) is then:
\begin{equation}
\sum F_x = \Big(P(x+dx)dydz -P(x)dydz \Big) + \Big( \tau (y+dy)dxdz - \tau (y)dxdz \Big) = ma_x
\end{equation}
We are interested in the flow at its steady state, when $ a_x = 0 $. We may devide the equation by $ dxdydz $, and find:
\begin{equation}
-\frac{dP}{dx} = \frac{d\tau}{dy} = \mu \frac{d^2 v_x}{dy^2}
\end{equation}
In wich we might solve for the velocity profile, and find:
\begin{equation}
v_x(y) = -\frac{\Delta P}{\mu L}\big( y^2 + Cy + D \big)
\end{equation}
In this equation we have to insert the boundary conditions.

Inserting water into a box using the fix pour command. Each timestep particles are inserted, they are placed randomly inside the insertion volume so as to mimic a stream of poured particles. If they are molecules they are also oriented randomly. Each atom in the particle is tested for overlaps with existing particles, including effects due to periodic boundary conditions if applicable. If an overlap is detected, another random insertion attempt is made; see the vol keyword discussion below. The larger the volume of the insertion region, the more particles that can be inserted at any one timestep. Particles are inserted again after enough time has elapsed that the previously inserted particles fall out of the insertion volume under the influence of gravity. Insertions continue every so many timesteps until the desired # of particles has been inserted.

At the same time we have to remove the same amount of particles at the other side of the pipe/tube to keep the number of particles in the system constant. For this we use the fix evaporate command. These commands are only available through the MISC and GRANULAR package which LAMMPS then has to be build with. The MISC package contains the fix evaporate command, and the GRANULAR package contains the fix pour package.
