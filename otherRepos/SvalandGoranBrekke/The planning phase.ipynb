### CO_2 - using the EMP2 model for the Carbon Dioxide force field

http://paros.princeton.edu/jerring/gibbs_old/results/co2_emp2/co2_emp2.html
Also refering to the article "Atomistice Molecular Dynamics Simulations of $CO_2$ Diffusivity in $H_2O$ for a Wide Range of Temperatures and Pressures" by Moultos et.al.
IN these datasets, the energy is given in Kelvin, from $ \epsilon /k_B $, where $ k_B $ is Bolzmanns constant. But in LAMMPS we have to give $ \epsilon $ in $ [Kcal/mol] $. $ \epsilon $ is given in Joules, and is $ [Joule] = [Kcal/4184]$. Then it is only to divide by Avogadros number to get $ \epsilon $ in $ [Kcal/mol] $. So, $ \epsilon = \frac{\epsilon}{k_B} \cdot k_B \cdot \frac{1}{4184 \cdot N_A} [\frac{Kcal}{mol}] $. 

Initializing the $ CO_2 $ we assume that at $ 1atm $ and at $ 0^oC $, then the density $ \rho = 1.977kg/m^3$. If we want to create a system of $ 20^3 = 8000 $ particles, this is $ 5.84622E-19 g $ ($ CO_2 $ has a molecular mass of $ 44.0095g/mol $), we easly calculate that the system size has to be $ 666.22 Aangstrom $, which leave a spacing between every atom by $ 33.31 Aangstrom $.

## Tetrahedral Structure parameter for water

The tetrahedral structure parameter for water is a measure of the degree of tetrahedral formation of the water molecules in the system. Water in its crystalline form tends to form this tetrahedral structure. Therefore, a way to calculate how thetrahedral the structure of the water is, would be a way of measureing how close the water is to be ice. In the figure below you see the watermolecules in a tetrahedral structure where the oxygen atoms are forming the structure. The center oxygen is atom number \textit{k}.

```python
from IPython.display import Image
Image(filename='/home/goran/Goran/Master/water/tetrahedral_water_angle.png')
```

The tetrahedral order parameter is defined as:
\begin{equation}
Q_k = 1 - \frac{3}{8}\sum _i^3\sum _{j=i+1}^4 \Big( cos(\theta _{ikj}) + \frac{1}{3} \Big) ^2
\end{equation}
Where the angle $ \theta _{ikj}$ is the angle between atom \textit{i} and \textit{j} for the structure parameter $ Q_k $ for atom \textit{k}. As the equation for the tetrahedral order parameter shows, we have to find all angles within the tetrahedron that is spanned from the center atom \textit{k} to atom \textit{i} and \textit{j}.

## Young's modulus

Young's modulus \textit{E}, is a way to measure the stiffness of an elastic material and is found by taking the ratio of the tensile stress and the extensional strain. 
\begin{equation}
E = \frac{\sigma }{\epsilon }
\end{equation}
Tensile stress is the net internal force, parallell to the direction of the strain, per unit area. To calculate the atomic stress see RRahman article for how LAMMPS does it...
